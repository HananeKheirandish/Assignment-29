```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/rubix.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
row, col, _ = img.shape

R, G, B = cv2.split(img)
# print(R)

for i in range(row):
    for j in range(col):
        if R[i, j]>120 and G[i, j]>120 and B[i, j]<120:
            img[i, j] = (0, 0, 255)
            
        elif R[i, j]>120 and G[i, j]<120 and B[i, j]>120:
            img[i, j] = (0, 255, 0)
            
        elif R[i, j]<120 and G[i, j]>120 and B[i, j]>120:
            img[i, j] = (255, 0, 0)
            
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2.imwrite('rubix.jpg', img)
```
