```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageDraw, ImageFont
```

```python
logo = np.ones((600, 1250, 3), dtype='uint8')

logo[:, :] = (90,90,90)

plt. imshow(logo)
```

```python
#Red
logo[195:295, 200:300] = (242,80,34)

#Blue
logo[305:405, 200:300] = (0,164,239)

#Green
logo[195:295, 310:410] = (127,186,0)

#Yellow
logo[305:405, 310:410] = (255,185,0)

plt.imshow(logo)
plt.show()
```

```python
#text
font = ImageFont.truetype("arial.ttf", 150)

logo_pil = Image.fromarray(logo)
draw = ImageDraw.Draw(logo_pil)
text = "Microsoft"
draw.text((450, 220), text, font=font, fill=(255, 255, 255))
logo = np.array(logo_pil)

plt.imshow(logo)
```

```python
logo = cv2.cvtColor(logo, cv2.COLOR_RGB2BGR)

cv2.imwrite('logo.jpg',logo)
```
